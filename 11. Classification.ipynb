$\large{\textbf{11.1 Classification}}$

*   In classification problems the labels are from a $\textbf{discrete}$ set with finite elements.
*   The objective is to find a model or function $f$ which can predict the class of the input data.

* Classification algorthms can be grouped as
   
   * $\color{red}{\textbf{Linear}}$: which classify the data based on the output value of a linear function.
   * $\color{red}{\textbf{Nonlinear}}$: which classify the data based on the output value of a nonlinear function.





$\large{\textbf{11.2 Illustrating Linear Classification}}$

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.RandomState(1)
a=3
b=2

Xtr1_1 = np.linspace(-6, 2, 50)
Xtr1_2 = np.linspace(-2, 6, 50)

Xtr2_1= a*Xtr1_1+5*b + 3*rng.randn(50)
Xtr2_2= a*Xtr1_2-5*b - 3*rng.randn(50)

ax = plt.axes()
ax.scatter(Xtr1_1, Xtr2_1, c = 'red');
ax.scatter(Xtr1_2, Xtr2_2, c = 'green');
plt.title("Linear Classification")
```

$\large{\textbf{11.3 Illustrating Nonlinear Classification}}$

```python
rng = np.random.RandomState(1)
a1=0.5; a2=0.1; a3=0.02; b=2;

Xtr1_1 = np.linspace(-8, 6, 100)
Xtr1_2 = np.linspace(-6, 8, 100)

Xtr2_1 = a3 * np.power(Xtr1_1,3) + a2 * np.power(Xtr1_1,2) + a1 * Xtr1_1 + 5*b + 3*rng.randn(100)
Xtr2_2 = a3 * np.power(Xtr1_2,3) + a2 * np.power(Xtr1_2,2) + a1 * Xtr1_2 - 3*b - 4*rng.randn(100)

ax = plt.axes()
ax.scatter(Xtr1_1, Xtr2_1, c = 'red');
ax.scatter(Xtr1_2, Xtr2_2, c = 'green');
plt.title("Nonlinear Classification")
```
