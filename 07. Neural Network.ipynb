$\large{\textbf{7.1 Neural Network (NN)}}$  

*  $\color{red}{\textbf{NN}}$ is a $\textbf{composite function}$ of the following form:
\begin{equation}
    \hat{y}=f_{L+1}(f_{L}(...(f_{1}(f_{0}(x))))
\end{equation}
where $f_{0},f_{1},f_{2},...,f_{L+1}$ are $\textbf{layers}$ of the NN which are composite functions of the form:
\begin{equation}
    f_{i}(x)=\sigma(Ax+b)
\end{equation}
*  The functions $f_{i}$ consist of two components:

    *  $\color{red}{\textbf{Affine function}}$: of the form $Ax+b.$

    *  $\color{red}{\textbf{Activation function}}$: can be of different types:
 1.  $\color{red}{\textbf{Linear}}$ function:
      \begin{equation}
        \sigma(x)=ax
      \end{equation}
 2. $\color{red}{\textbf{Step}}$ function:
     \begin{equation}
        \sigma(x)=\begin{cases}
        0, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        1, \hspace{0.3cm} &if \hspace{0.1cm} x >0
        \end{cases}
     \end{equation}
 3. $\color{red}{\textbf{ReLU}}$ (Rectified Linear Unit) function:
    \begin{equation}
        \sigma(x)=\max(0,x)= \begin{cases}
        0, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        x, \hspace{0.3cm} &if \hspace{0.1cm} x >0
        \end{cases}
    \end{equation}
 4. $\color{red}{\textbf{Sigmoid}}$ function:
    \begin{equation}
      \sigma(x)=\frac{1}{1+e^{-x}}
    \end{equation}
 5.   $\color{red}{\textbf{Tanh}}$ (hyperbolic tan) function:
    \begin{equation}
        \sigma(x)=\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}
    \end{equation}

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-10, 10, 100)
Y=np.zeros(len(X))
for i in range(len(X)):
 Y[i] = np.max([(X[i]),0])
plt.plot(X, Y, c = 'red');
```

```python
X = np.linspace(-10, 10, 100)
Y=np.zeros(len(X))
for i in range(len(X)):
 Y[i] = np.max([(2*X[i]+1),0])
plt.plot(X, Y, c = 'red');
```

```python
X = np.linspace(-10, 10, 100)
Y=np.zeros(len(X))
for i in range(len(X)):
 Y[i] = np.max([(-X[i]-2),0])
plt.plot(X, Y, c = 'red');
```

```python
X = np.linspace(-10, 10, 100)
Y1=np.zeros(len(X))
Y2=np.zeros(len(X))
Y3=np.zeros(len(X))
Y=np.zeros(len(X))
for i in range(len(X)):
 Y1[i] = np.max([(2*X[i]+1),0])
 Y2[i] = np.max([(-0.1*Y1[i]+1.5),0])
 Y3[i] = np.max([(-0.3*Y2[i]+.5),0])
 Y[i]  = Y3[i]
plt.plot(X, Y, c = 'red');
```

```python
X = np.linspace(-10, 10, 100)
Y1=np.zeros(len(X))
Y2=np.zeros(len(X))
Y3=np.zeros(len(X))
Y=np.zeros(len(X))
for i in range(len(X)):
 Y1[i] = np.max([(2*X[i]+1),0])
 Y2[i] = np.max([(-0.1*Y1[i]+1.5),0])
 Y3[i] = np.max([(-0.3*Y2[i]+.5),0])
 Y[i]  = Y3[i]
fig, axs = plt.subplots(1, 3, figsize =(12, 4))
axs[0].plot(X, 2*X+1, c = 'r');
axs[1].plot(X, Y1, c = 'r')
axs[2].plot(X, Y, c = 'r');
```

```python
X = np.linspace(-10, 10, 100)
Y11=np.zeros(len(X))
Y12=np.zeros(len(X))
Y=np.zeros(len(X))
for i in range(len(X)):
 Y11[i] = np.max([(2*X[i]+1),0])
 Y12[i] = np.max([(-1*X[i]-2),0])
 Y[i]  = Y11[i]+Y12[i]
plt.plot(X, Y, c = 'red');
```

```python
X = np.linspace(-10, 10, 100)
Y11=np.zeros(len(X))
Y12=np.zeros(len(X))
Y13=np.zeros(len(X))
Y=np.zeros(len(X))
for i in range(len(X)):
 Y11[i] = np.max([(3*X[i]+1),0])
 Y12[i] = np.max([(-1.5*X[i]+1.5),0])
 Y13[i] = np.max([(-0.5*X[i]-1.5),0])
 Y[i]  = Y11[i]+Y12[i]+Y13[i]
plt.plot(X, Y, c = 'red');
```
