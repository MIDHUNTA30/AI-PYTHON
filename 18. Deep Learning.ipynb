$\large{\textbf{18.1 Deep Learning (DL)}}$

* $\color{red}{\textbf{DL}}$  deals with deep NN-based modeling and is a subfield of NN.


* Popular models in DL:
   * $\color{red}{\textbf{Feedforward Neural Network}}$ (FNN): in which the layers are connected sequentially and used to model static relationships.  
   * $\color{red}{\textbf{Autoencoder}}$ (AE): consists of two sequentially connected FNNs trained together to reconstruct the input data and used for nonlinear dimensionality reduction.
   * $\color{red}{\textbf{Convolutional Neural Network}}$ (CNN): uses convolutional filters to model sequential relationships in the data.
   * $\color{red}{\textbf{Recurrent Neural Network}}$ (RNN): used to model dynamic relationships between input-output data.
   * $\color{red}{\textbf{Transformer}}$: uses attention mechanism to extract dynamic relationships in the data.

* Applications of DL:

  * Computer vision: deals with problems such as image recognition, image classification.
  * Process modeling: deals with learning static and dynamic relationships amongst variables in the process data.
  * Medical applications: such as medical diagnosis, drug discovery
  * Finance: such as sales prediction, profit analysis

$\large{\textbf{18.2 Datasets for Image Classification}}$

```python
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.datasets import cifar10
```

```python
# Load MNIST dataset
(Xtr, Ytr), (Xts, Yts) = mnist.load_data()
```

```python
fig, axes = plt.subplots(2, 5, figsize=(10, 4))
for i, ax in enumerate(axes.flat):
    ax.imshow(Xtr[i], cmap='gray')
    ax.set_title(f"Label: {Ytr[i]}")
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Load CIFAR-10 dataset
(Xtr, Ytr), (Xts, Yts) = cifar10.load_data()
```

```python
fig, axes = plt.subplots(2, 5, figsize=(10, 4))
for i, ax in enumerate(axes.flat):
    ax.imshow(Xtr[i])
    ax.set_title(f"Label: {Ytr[i][0]}")
    ax.axis('off')
plt.tight_layout()
plt.show()
```
