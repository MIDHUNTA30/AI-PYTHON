$\large{\textbf{9.1 Neural Network Regression}}$

*   Neural Network (NN) regression uses NN model:
\begin{equation}
    \hat{y}=f_{NN}(x,P)=f_{L+1}(f_{L}(...(f_{0}(x))))
\end{equation}
where $f_{0},f_{1},...,f_{L+1}$ are layers of the NN.
*   The parameter $P$ is the weights and biases of each layer  in the NN which are to be optimized.
* To find $P$ various optimization algorithms can be used that uses the training data and computes the optimal parameters which minimizes the cost function $J(y_{tr}-\hat{y})$.

$\large{\textbf{9.2 NN Regression Example}}$

* Training input: $\{x_{1},...,x_{20} \}$ spaced uniformly in range $[0, 1]$.
* The output data is generated as
\begin{equation}
y_{tr}=e^{2x_{tr}}+0.2 r
\end{equation}
where $r$ is a Gaussian random variable with mean 0 and variance 1.

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
```

```python
rng = np.random.RandomState(1)
X = np.linspace(0, 1, 100)
Y = np.exp(2*X)

rng = np.random.RandomState(1)
Xtr = np.linspace(0, 1, 20)
Ytr = np.exp(2*Xtr) + 0.2*rng.randn(len(Xtr))

plt.figure(figsize =(6, 4))
ax = plt.axes()
plt.scatter(Xtr, Ytr,  c = 'r');
plt.plot(X, Y, c = 'k');
```

```python
f_NN = MLPRegressor(hidden_layer_sizes=(3,3,3),activation="relu" , solver='lbfgs', max_iter=10000)
#lbfgs:  Limited-memory Broyden-Fletcher-Goldfarb-Shanno    # for displaying error use verbose=True
f_NN.fit(Xtr[:, np.newaxis], Ytr)
YhatNN = f_NN.predict(X[:, np.newaxis])

plt.figure(figsize =(6, 4))
plt.scatter(Xtr, Ytr, c = 'r')
plt.plot(X, Y, c = 'k');
plt.plot(X, YhatNN, c = 'g');
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.legend(['$y_{tr}$','$y$', '$\hat{y}_{NN}$'])
plt.show()
```
