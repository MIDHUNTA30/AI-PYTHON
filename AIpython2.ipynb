$\large{\textbf{2.1 About Python}}$
*   $\color{red}{\textbf{Python}}$ is a $\textbf{high-level}$ programming language.

$\large{\textbf{2.2 Terms and Definitions}}$
1. $\color{red}{\textbf{function}}$ (in programming): a block of code that performs a specific task.
2. $\color{red}{\textbf{class}}$: a collection functions.
2. $\color{red}{\textbf{module}}$: a collection classes.
3. $\color{red}{\textbf{package}}$: a collection of modules.
3. $\color{red}{\textbf{library}}$: a collection of packages.


$\large{\textbf{2.3 Useful Python libraries for AI}}$
1. $\color{red}{\textbf{numpy}}$ (Numeric Python): contains functions to perform mathematical operations on arrays and matrices.
2. $\color{red}{\textbf{matplotlib}}$: contains functions for plotting and visualization.
3. $\color{red}{\textbf{scipy}}$ (Scientific Python): contains functions for optimization, linear algebra, calculus, signal and image processing, etc.
4. $\color{red}{\textbf{pandas}}$ (Python data analysis): contains functions preprocessing, analyzing and manipulating data etc.
5. $\color{red}{\textbf{seaborn}}$: is a data visualization library in Python. Contains functions for statistical graphics.
6. $\color{red}{\textbf{sklearn}}$ (Scikit-Learn): used for implementing machine learning models. Contains functions for regression, classification, etc.
7. $\color{red}{\textbf{keras}}$: contains functions for implementing neural networks and deep learning algorithms. Normally used for small data sets.
8. $\color{red}{\textbf{tensorflow}}$: contains functions for implementing neural networks and deep learning algorithms. Normally used for large data sets.
9. $\color{red}{\textbf{pytorch}}$: contains functions for implementing machine learning and deep learning algorithms.
10. $\color{red}{\textbf{nltk}}$ (Natural Language ToolKit): contains functions for natural language processing.

$\large{\textbf{2.8 Loops in Python}}$

```python
a=3
b=2
X = np.linspace(0, 10, 100)
Y=np.zeros(len(X))
for i in range(len(X)):
 Y[i] = a * X[i] + b
plt.scatter(X, Y, c = 'red');
```

```python
a=3
b=2
X = np.linspace(0, 10, 100)
Y=np.zeros(len(X))
i = 0
while i < len(X):
 Y[i] = a * X[i] + b
 i += 1
plt.scatter(X, Y, c = 'red');
```
