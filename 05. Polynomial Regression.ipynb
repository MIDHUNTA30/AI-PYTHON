$\large{\textbf{5.1 Polynomial Regression}}$  

*  $\color{red}{\textbf{Polynomial regression}}$ is used to model $\textbf{nonlinear}$ input-output relationships.

* $\color{red}{\textbf{Model}}$ is a $\textbf{polynomial}$ function:
  \begin{equation}
      \hat{y}= f(x) =p_{M}x^{M}+...+p_{1}x+p_{0}
  \end{equation}

*   Training data contains

 1.   $\color{red}{\textbf{Inputs}}$: consists of $N$ observations $\{ x_{1}, x_{2},...,x_{N} \}$.
 2.   $\color{red}{\textbf{Outputs}}$: consists of $N$ labels $\{y_{1}, y_{2}, ..., y_{N} \}$.  

$\large{\textbf{5.2 Polynomial Regression: Example 1}}$

* Training input: $\{x_{1},...,x_{10} \}$ spaced uniformly in range $[0, 1]$.
* The training output is generated as
\begin{equation}
y_{tr}=0.02x_{tr}^{3}+0.1x_{tr}^{2}+0.5x_{tr}+2+r
\end{equation}
where $r$ is a Gaussian random variable with mean 0 and variance 1.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a1=0.5; a2=0.1; a3=0.02; b=2;
X = np.linspace(-10, 10, 100)
Y = a3 * np.power(X,3) + a2 * np.power(X,2) + a1 * X + b

rng = np.random.RandomState(1)
Xtr = np.linspace(-10, 10, 20)
Ytr = a3 * np.power(Xtr,3) + a2 * np.power(Xtr,2) + a1 * Xtr + b +  rng.randn(len(Xtr))
plt.scatter(Xtr, Ytr,  c = 'r');
plt.plot(X, Y, c = 'k');
```

```python
model_P = np.poly1d(np.polyfit(Xtr, Ytr, 3))
Y_P = model_P(X[:, np.newaxis])
plt.scatter(Xtr, Ytr, c = 'r')
plt.plot(X, Y, c = 'k');
plt.plot(X, Y_P, c = 'g');
```

```python
np.polyfit(Xtr, Ytr, 3)
```

```python
print(model_P)
```

$\large{\textbf{5.3 Polynomial Regression: Example 2}}$

* Training input: $\{x_{1},...,x_{10} \}$ spaced uniformly in range $[0, 1]$.
* The training output is generated as
\begin{equation}
y_{tr}=e^{2x_{tr}}+0.5 r
\end{equation}
where $r$ is a Gaussian random variable with mean 0 and variance 1.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 1, 100)
Y = np.exp(2*X)

rng = np.random.RandomState(1)
Xtr = np.linspace(0, 1, 10)
Ytr = np.exp(2*Xtr) + 0.5*rng.randn(len(Xtr))

plt.scatter(Xtr, Ytr,  c = 'r');
plt.plot(X, Y, c = 'k');
```

```python
model_P = np.poly1d(np.polyfit(Xtr, Ytr, 3))
Y_P = model_P(X[:, np.newaxis])
plt.scatter(Xtr, Ytr, c = 'r')
plt.plot(X, Y, c = 'k');
plt.plot(X, Y_P, c = 'g');
```

```python
np.polyfit(Xtr, Ytr, 3)
```

```python
print(model_P)
```

${\textbf{Effect of model order}}$

```python
model_P0 = np.poly1d(np.polyfit(Xtr, Ytr, 0))
model_P1 = np.poly1d(np.polyfit(Xtr, Ytr, 1))
model_P2 = np.poly1d(np.polyfit(Xtr, Ytr, 2))
model_P3 = np.poly1d(np.polyfit(Xtr, Ytr, 3))
model_P4 = np.poly1d(np.polyfit(Xtr, Ytr, 4))
model_P5 = np.poly1d(np.polyfit(Xtr, Ytr, 5))
model_P6 = np.poly1d(np.polyfit(Xtr, Ytr, 6))
model_P7 = np.poly1d(np.polyfit(Xtr, Ytr, 7))
model_P8 = np.poly1d(np.polyfit(Xtr, Ytr, 8))
model_P9 = np.poly1d(np.polyfit(Xtr, Ytr, 9))

Y_P0 = model_P0(X[:, np.newaxis])
Y_P1 = model_P1(X[:, np.newaxis])
Y_P2 = model_P2(X[:, np.newaxis])
Y_P3 = model_P3(X[:, np.newaxis])
Y_P4 = model_P4(X[:, np.newaxis])
Y_P5 = model_P5(X[:, np.newaxis])
Y_P6 = model_P6(X[:, np.newaxis])
Y_P7 = model_P7(X[:, np.newaxis])
Y_P8 = model_P8(X[:, np.newaxis])
Y_P9 = model_P9(X[:, np.newaxis])
```

```python
fig, axs = plt.subplots(2, 5, figsize =(12, 6))
axs[0, 0].scatter(Xtr, Ytr, c = 'r')
axs[0, 0].plot(X, Y, c = 'k');
axs[0, 0].plot(X, Y_P0, c = 'g');
axs[0, 1].scatter(Xtr, Ytr, c = 'r')
axs[0, 1].plot(X, Y, c = 'k');
axs[0, 1].plot(X, Y_P1, c = 'g');
axs[0, 2].scatter(Xtr, Ytr, c = 'r')
axs[0, 2].plot(X, Y, c = 'k');
axs[0, 2].plot(X, Y_P2, c = 'g');
axs[0, 3].scatter(Xtr, Ytr, c = 'r')
axs[0, 3].plot(X, Y, c = 'k');
axs[0, 3].plot(X, Y_P3, c = 'g');
axs[0, 4].scatter(Xtr, Ytr, c = 'r')
axs[0, 4].plot(X, Y, c = 'k');
axs[0, 4].plot(X, Y_P4, c = 'g');
axs[1, 0].scatter(Xtr, Ytr, c = 'r')
axs[1, 0].plot(X, Y, c = 'k');
axs[1, 0].plot(X, Y_P5, c = 'g');
axs[1, 1].scatter(Xtr, Ytr, c = 'r')
axs[1, 1].plot(X, Y, c = 'k');
axs[1, 1].plot(X, Y_P6, c = 'g');
axs[1, 2].scatter(Xtr, Ytr, c = 'r')
axs[1, 2].plot(X, Y, c = 'k');
axs[1, 2].plot(X, Y_P7, c = 'g');
axs[1, 3].scatter(Xtr, Ytr, c = 'r')
axs[1, 3].plot(X, Y, c = 'k');
axs[1, 3].plot(X, Y_P8, c = 'g');
axs[1, 4].scatter(Xtr, Ytr, c = 'r')
axs[1, 4].plot(X, Y, c = 'k');
axs[1, 4].plot(X, Y_P9, c = 'g');
```

${\textbf{Error analysis}}$

```python
from sklearn.metrics import mean_squared_error

Xts = np.linspace(0.05, 1.05, 10)
Yts = np.exp(2*Xts)+ 0.5*rng.randn(len(Xts))

Y_Pr0 = model_P0(Xtr[:, np.newaxis])
Y_Pr1 = model_P1(Xtr[:, np.newaxis])
Y_Pr2 = model_P2(Xtr[:, np.newaxis])
Y_Pr3 = model_P3(Xtr[:, np.newaxis])
Y_Pr4 = model_P4(Xtr[:, np.newaxis])
Y_Pr5 = model_P5(Xtr[:, np.newaxis])
Y_Pr6 = model_P6(Xtr[:, np.newaxis])
Y_Pr7 = model_P7(Xtr[:, np.newaxis])
Y_Pr8 = model_P8(Xtr[:, np.newaxis])
Y_Pr9 = model_P9(Xtr[:, np.newaxis])

Y_Ps0 = model_P0(Xts[:, np.newaxis])
Y_Ps1 = model_P1(Xts[:, np.newaxis])
Y_Ps2 = model_P2(Xts[:, np.newaxis])
Y_Ps3 = model_P3(Xts[:, np.newaxis])
Y_Ps4 = model_P4(Xts[:, np.newaxis])
Y_Ps5 = model_P5(Xts[:, np.newaxis])
Y_Ps6 = model_P6(Xts[:, np.newaxis])
Y_Ps7 = model_P7(Xts[:, np.newaxis])
Y_Ps8 = model_P8(Xts[:, np.newaxis])
Y_Ps9 = model_P9(Xts[:, np.newaxis])

mse_tr = np.linspace(0, 1, 10)
mse_tr[0] = mean_squared_error(Ytr,Y_Pr0)
mse_tr[1] = mean_squared_error(Ytr,Y_Pr1)
mse_tr[2] = mean_squared_error(Ytr,Y_Pr2)
mse_tr[3] = mean_squared_error(Ytr,Y_Pr3)
mse_tr[4] = mean_squared_error(Ytr,Y_Pr4)
mse_tr[5] = mean_squared_error(Ytr,Y_Pr5)
mse_tr[6] = mean_squared_error(Ytr,Y_Pr6)
mse_tr[7] = mean_squared_error(Ytr,Y_Pr7)
mse_tr[8] = mean_squared_error(Ytr,Y_Pr8)
mse_tr[9] = mean_squared_error(Ytr,Y_Pr9)

mse_ts = np.linspace(0, 1, 10)
mse_ts[0] = mean_squared_error(Yts,Y_Ps0)
mse_ts[1] = mean_squared_error(Yts,Y_Ps1)
mse_ts[2] = mean_squared_error(Yts,Y_Ps2)
mse_ts[3] = mean_squared_error(Yts,Y_Ps3)
mse_ts[4] = mean_squared_error(Yts,Y_Ps4)
mse_ts[5] = mean_squared_error(Yts,Y_Ps5)
mse_ts[6] = mean_squared_error(Yts,Y_Ps6)
mse_ts[7] = mean_squared_error(Yts,Y_Ps7)
mse_ts[8] = mean_squared_error(Yts,Y_Ps8)
mse_ts[9] = mean_squared_error(Yts,Y_Ps9)

fig, axs = plt.subplots(1, 3, figsize =(12, 4))
axs[0].plot(mse_tr, c = 'red')
axs[0].set_xlabel('M')
axs[0].set_ylabel('mse')
axs[0].set_title("MSE (Training)")
axs[1].plot(mse_ts, c = 'green')
axs[1].set_xlabel('M')
axs[1].set_ylabel('mse')
axs[1].set_title("MSE (Testing)")
axs[2].plot(mse_tr, c = 'red')
axs[2].plot(mse_ts, c = 'green')
axs[2].set_xlabel('M')
axs[2].set_ylabel('mse')
axs[2].set_title("MSE Comparison")
```
