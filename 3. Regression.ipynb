$\large{\textbf{3.1 Supervised Learning}}$

*   Supervised Learning is used for obtaining model of a set of data containing the inputs and outputs ($\textbf{labels}$).
*   Based on the nature or output or label or target data, AI algorithms can be grouped as
      1. $\color{red}{\textbf{Regression}}$: in which outputs or labels are continuous variables.
      1. $\color{red}{\textbf{Classification}}$: in which outputs or labels are discrete and finite in number.



$\large{\textbf{3.2 Terms and Definitions}}$
1.  Input or observation: $x$
2.  Output or label: $y$
3.  Input data or input matrix: $X$
4.  Output data or output (label) matrix: $Y$
5.  Training input data: $X_{tr}$
6.  Training output data: $Y_{tr}$
7.  Validation input data: $X_{va}$
8.  Validation output data: $Y_{va}$
9.  Testing input data: $X_{ts}$
10. Testing output data: $Y_{ts}$
11. Univariate data: consists of observations of only one variable.
12. Multivariate data: consists of observations of more than one variables.

$\large{\textbf{3.3 Regression}}$


*   In regression problems the outputs or labels are from a $\textbf{continuous}$ set.
*   The objective is to find a model or function $f$ which map the input feature vector to a continuous output:
\begin{equation}
\hat{y} = f(P,x)
\end{equation}

*   The training data consists of

 1.   $\color{red}{\textbf{Inputs}}$: consists of $N$ observations $\{ x_{1}, x_{2},...,x_{N} \}$.
 2.   $\color{red}{\textbf{Outputs}}$: consists of $N$ labels $\{y_{1}, y_{2}, ..., y_{N} \}$.

*   Model is a function:

 *   $f : \mathbb{R} → \mathbb{R}$ in univariate case.
 *   $f : \mathbb{R}^{n} → \mathbb{R}^{m}$ in multi-variate case.


*   Regression algorithms can be grouped as
     1. $\color{red}{\textbf{Linear}}$: in which the function $f$ is linear.
     3. $\color{red}{\textbf{Nonlinear}}$: in which the function $f$ is nonlinear.
      





$\large{\textbf{3.4 Illustrating Linear Regression}}$

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=3
b=2
X = np.linspace(-10, 10, 100)
Y = a*X+b

rng = np.random.RandomState(1)
Xtr = np.linspace(-10, 10, 20)
Ytr = a * Xtr + b + 2* rng.randn(len(Xtr))

ax = plt.axes()
ax.scatter(Xtr, Ytr, c = 'red');
ax.plot(X, Y, c = 'black')
plt.title("Linear Regression")
```

$\large{\textbf{3.4 Illustrating Nonlinear Regression}}$

```python
a1=0.5; a2=0.1; a3=0.02; b=2;
X = np.linspace(-10, 10, 100)
Y= a3 * np.power(X,3) + a2 * np.power(X,2) + a1 * X + b

rng = np.random.RandomState(1)
Xtr = np.linspace(-10, 10, 20)
Ytr = a3 * np.power(Xtr,3) + a2 * np.power(Xtr,2) + a1 * Xtr + b + rng.randn(len(Xtr))

ax = plt.axes()
ax.scatter(Xtr, Ytr, c = 'r');
ax.plot(X, Y, c = 'k')
plt.title("Nonlinear Regression")
```
