$\large{\textbf{14.1 Dimensionality Reduction}}$

* In general,
real-world high dimensional data is likely to concentrate in the vicinity of lower dimensional manifolds.

* This resulted in the subfield of unsupervised learning known as $\color{red}{\textbf{dimensionality reduction}}$.

*  Two major problems of interest in dimensionality reduction are:
      *  $\color{red}{\textbf{Feature extraction}}$: where the objective is to extract lower dimensional features.
      *  $\color{red}{\textbf{Model extraction}}$: deals with identifying the relationships amongst the input variables.
* Dimensionality reduction methods:
     * $\color{red}{\textbf{Principal Component Analysis (PCA)}}$: is a linear dimensionality reduction method.
     * $\color{red}{\textbf{Autoencoder}}$: is a nonlinear dimensionality reduction methods.


$\large{\textbf{14.2 Generating dataset for Dimensionality Reduction}}$

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import seaborn as sns
```

```python
N=100;
# Generate random data for the first 4 variables
X = np.random.rand(N, 4)
# Linear relationships
x5 = 2*X[:, 0] + 3*X[:, 1] - X[:, 2] + np.random.normal(0, 0.05, N)
x6 = -X[:, 3] + 0.5*X[:, 2] + 2*X[:, 1] + np.random.normal(0, 0.05, N)
x7 = 2*X[:, 2] + np.random.normal(0, 0.05, N)
x8 = 0.7 * x5 + 0.3 * x6 + np.random.normal(0, 0.001, N)
# Non-linear relationships
x9 = np.sin(2 * np.pi * X[:, 0]) + np.cos(np.pi * X[:, 1]) + np.random.normal(0, 0.01, N)
x10 = np.cos(np.pi * x7) + np.random.normal(0, 0.01, N)
# Training data
Xtr = np.column_stack((X, x5, x6, x7, x8, x9, x10))
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(Xtr[:, 2], Xtr[:, 6], c = 'red')
plt.xlabel("x3")
plt.ylabel("x7")
plt.title("Plot of x3 vs x7")
plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(Xtr[:, 4], Xtr[:, 5], Xtr[:, 7], c = 'red')
ax.set_xlabel("x5")
ax.set_ylabel("x6")
ax.set_zlabel("x8")
ax.set_title("Plot of x5 vs x6 vs x8")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(Xtr[:, 6], Xtr[:, 9], c = 'red')
plt.xlabel("x7")
plt.ylabel("x10")
plt.title("Plot of x7 vs x10")
plt.show()
```

```python
sns.pairplot(data=pd.DataFrame(Xtr))
plt.show()
```
