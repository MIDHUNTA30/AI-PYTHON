$\large{\textbf{4.1 Linear Regression}}$  


*    $\color{red}{\textbf{Linear regression}}$ is used to fit the input-output data with a linear function $f$:
\begin{equation}
\hat{y} = f(x) = Ax + b
\end{equation}
*    The task is to find the $\textbf{optimal}$ weight $A$ and bias $b$ for a given data.

*   Training data contains

 1.   $\color{red}{\textbf{Inputs}}$: consists of $N$ observations $\{ x_{1}, x_{2},...,x_{N} \}$.
 2.   $\color{red}{\textbf{Outputs}}$: consists of $N$ labels $\{y_{1}, y_{2}, ..., y_{N} \}$.

* $\color{red}{\textbf{Model}}$ is a $\textbf{linear}$ function:

  *   $f : \mathbb{R} → \mathbb{R}$ in univariate case.
  *   $f : \mathbb{R}^{n} → \mathbb{R}^{m}$ in multi-variate case.





$\large{\textbf{4.2 Linear Regression Example}}$

* Training input: $\{x_{1},...,x_{100} \}$ spaced uniformly in range $[0, 10]$.
* The training data is generated as
\begin{equation}
y_{tr}=3x_{tr}+2 + r
\end{equation}
where $r$ is a Gaussian random variable with mean 0 and variance 1.

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
a=3
b=2
X = np.linspace(0, 10, 1000)
Y = a * X + b

rng = np.random.RandomState(1)
Xtr = np.linspace(0, 10, 100)
Ytr = a * Xtr + b + rng.randn(len(Xtr))
plt.scatter(Xtr, Ytr, c = 'red');
```

```python
# create a linear regression model
model = LinearRegression()
model.fit(Xtr[:, np.newaxis], Ytr)
Y_L = model.predict(X[:, np.newaxis])

# plotting results
plt.figure(figsize =(7, 5))
ax = plt.axes()
ax.scatter(Xtr, Ytr, c = 'red')
ax.plot(X, Y, c = 'black')
ax.plot(X, Y_L, c = 'green')
#plt.grid()
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.legend(['$y_{tr}$','$y$', '$\hat{y}_{L}$'])

plt.show()
```

```python
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```
