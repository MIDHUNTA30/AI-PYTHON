$\large{\textbf{13.1 Decision Tree}}$

* $\color{red}{\textbf{Decision Tree}}$ is a nonlinear supervised model.
* Decision tree can be used either as a regressor or classifier.
*  Trees with discrete output are called classification trees.  
*  Trees with continous output are called regression trees.
*   Training data contains:

   *  $\color{red}{\textbf{Inputs}}$: consists of $N$ observations $\{ x_{1}, x_{2},...,x_{N} \}$
   where $x_{k} \in \mathbb{R}^n,$ $k=1,..,N$ are the $N$ input samples.
   *  $\color{red}{\textbf{Outputs}}$: consists of $N$ observations $\{ y_{1}, y_{2},...,y_{N} \}$
   where $y_{k} \in \mathbb{R}$  are the labels.
* Decision trees consists of:
     * Branches: represents decisions on features or inputs.
     * Leafs: represent class (or output) labels.



* $\color{red}{\textbf{Decision Boundary}}$: is the dividing line or set of conditions at each node of the tree that separates the data into different regions.

$\large{\textbf{13.2 Illustrating Decision Tree}}$

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import graphviz
import pydot
from sklearn import tree
from sklearn.metrics import r2_score
```

```python
rng = np.random.RandomState(1)
a=3
b=2

Xtr1_1 = np.linspace(-6, 2, 50)
Xtr1_2 = np.linspace(-2, 6, 50)
Xtr1=np.concatenate((Xtr1_1,Xtr1_2),axis=0)

Xtr2_1= a*Xtr1_1+5*b + 3*rng.randn(50)
Xtr2_2= a*Xtr1_2-5*b - 3*rng.randn(50)
Xtr2=np.concatenate((Xtr2_1,Xtr2_2),axis=0)
Xtr=(Xtr1,Xtr2)
Ytr_1=-1*np.ones(50)
Ytr_2=np.ones(50)
Ytr=np.concatenate((Ytr_1,Ytr_2),axis=0)
ax = plt.axes()
ax.scatter(Xtr1_1, Xtr2_1, c = 'green');
ax.scatter(Xtr1_2, Xtr2_2, c = 'red');
plt.title("Training Input")
```

```python
distillDTclass = tree.DecisionTreeClassifier(criterion='gini', max_depth = 3)
distillDTclass
```

```python
distillDTclass.fit(np.transpose(Xtr), Ytr)
```

```python
dot_data = tree.export_graphviz(distillDTclass, out_file=None,
                                feature_names=['Xtr1', 'Xtr2'],
                                class_names=['Class 1', 'Class 2'],
                                filled=True, rounded=True,
                                special_characters=True)
graph = graphviz.Source(dot_data)
graph
```

```python
# Predict the class for the training data
Ytrhat = distillDTclass.predict(np.transpose(Xtr))

# Plot the training data and predicted output
plt.figure(figsize=(8, 6))
plt.scatter(Xtr1_1, Xtr2_1, c='green', label='Class 1 (Training Data)')
plt.scatter(Xtr1_2, Xtr2_2, c='red', label='Class 2 (Training Data)')

# Plot predicted classes
plt.scatter(Xtr1[Ytrhat == -1], Xtr2[Ytrhat == -1], marker='x', c='blue', s=50, label='Predicted Class 1')
plt.scatter(Xtr1[Ytrhat == 1], Xtr2[Ytrhat == 1], marker='x', c='purple', s=50, label='Predicted Class 2')

plt.title("Decision Tree Classification - Training Data and Predictions")
plt.xlabel("Xtr1")
plt.ylabel("Xtr2")
plt.legend()
plt.show()
```

```python
# Create a meshgrid for plotting the decision boundary
xx, yy = np.meshgrid(np.arange(min(Xtr1) - 1, max(Xtr1) + 1, 0.1),
                     np.arange(min(Xtr2) - 1, max(Xtr2) + 1, 0.1))

# Predict the class for each point in the meshgrid
Z = distillDTclass.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.4)

# Plot the training data
plt.scatter(Xtr1_1, Xtr2_1, c='green', label='Class 1 (Training Data)')
plt.scatter(Xtr1_2, Xtr2_2, c='red', label='Class 2 (Training Data)')


plt.title("Decision Tree Decision Boundary")
plt.xlabel("Xtr1")
plt.ylabel("Xtr2")
plt.legend()
plt.show()
```
