$\large{\textbf{15.1 Principal Component Analysis (PCA)}}$

* $\color{red}{\textbf{PCA}}$ is a linear dimensionality reduction method.

*   Training data contains

   *  $\color{red}{\textbf{Inputs}}$ (Unlabeled data): consists of $N$ observations $\{ x_{1}, x_{2},...,x_{N} \}$
   where $x_{k} \in \mathbb{R}^n$ $k=1,..,N$ are the $N$ samples which are assumed to be mean-centered.
* Each sample $x$ consists of measurements of $n$ variables.

* Dimensionality reduction is achieved in PCA using linear transformation:
\begin{equation}
  y=P^{\top}x
\end{equation}

where $x \in \mathbb{R}^{n}$, $y \in \mathbb{R}^{m}$, $ m \leq n,$  and $ P \in \mathbb{R}^{n\times m}.$

* From $y$ the reconstructed input is computed as
\begin{equation}
    \hat{x} = Py=PP^{\top}x.
\end{equation}

* Optimization problem for PCA:
\begin{equation}
\begin{aligned}
    \underset{P}{\min} \hspace{.2cm}   & \sum_{k=1}^{\text{N}}  {\parallel x_{k} - PP^{\top}x_{k} \parallel }^{2}\\
    \text{subj. to}  ~&~ P^{\top}P=I_{m}
\end{aligned}    
\end{equation}


$\large{\textbf{15.2 Illustrating PCA}}$

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
N=100;
# Generate random data for the first 4 variables
X = np.random.rand(N, 4)
# Linear relationships
x5 = 2*X[:, 0] + 3*X[:, 1] - X[:, 2] + np.random.normal(0, 0.05, N)
x6 = -X[:, 3] + 0.5*X[:, 2] + 2*X[:, 1] + np.random.normal(0, 0.05, N)
x7 = X[:, 1] - 3*X[:, 2] + np.random.normal(0, 0.05, N)
x8 = 0.7 * x5 + 0.3 * x6 + np.random.normal(0, 0.001, N)
# Non-linear relationships
x9 = np.sin(2 * np.pi * X[:, 0]) + np.cos(np.pi * X[:, 1]) + np.random.normal(0, 0.01, N)
x10 = np.cos(np.pi * x7) + np.random.normal(0, 0.01, N)
# The training data
Xtr = np.column_stack((X, x5, x6, x7, x8, x9, x10))
```

```python
m = 6
scaler = StandardScaler()
X=scaler.fit_transform(Xtr)  # normalized data
```

```python
pca = PCA(n_components=m)
pca.fit(X)
P = pca.components_
V = pca.explained_variance_
Y=np.dot(X,P.T)
Xhat=np.dot(Y,P)
```

```python
msePCA=mean_squared_error(X,Xhat)
msePCA
```

```python
ax = plt.axes()
ax.plot(X[:,1], c = 'red')
plt.plot(Xhat[:,1], c = 'green')
plt.xlabel('$k$', fontsize=14)
plt.ylabel('$x_{1}$', fontsize=14)
plt.legend(['$x_{1}$','$\hat{x}_{1}$'])
plt.show()
```
