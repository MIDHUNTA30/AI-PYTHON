$\large{\textbf{17.1 Clustering}}$

* $\color{red}{\textbf{Clustering}}$ is an unsupervised classification method.


* $\color{red}{\textbf{K-means clustering}}$: aims to group the samples into K clusters.

* K-means clustering minimizes within-cluster variances.

*   Training data contains

   *  $\color{red}{\textbf{Inputs}}$ (Unlabeled data): consists of $N$ observations $\{ x_{1}, x_{2},...,x_{N} \}$
   where $x_{k} \in \mathbb{R}^n$ $k=1,..,N$ are the $N$ samples which are assumed to be mean-centered.

*  K-means clustering aims to partition the N observations into $K â‰¤ N$ sets $S = \{S_{1}, S_{2}, ..., S_{K}\}$.

* Optimization problem for clustering:
\begin{equation}
    \underset{S}{\min} \hspace{.2cm} \sum_{i=1}^{K} |S_{i}|Var(S_{i})  
\end{equation}




$\large{\textbf{17.2 Illustrating K-means clustering}}$

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
rng = np.random.RandomState(1)
a = 3
b = 2

Xtr1_1 = np.linspace(-6, 2, 50)
Xtr1_2 = np.linspace(-2, 6, 50)
Xtr1 = np.concatenate((Xtr1_1, Xtr1_2), axis=0)
Xtr2_1 = a * Xtr1_1 + 5 * b + 3 * rng.randn(50)
Xtr2_2 = a * Xtr1_2 - 5 * b - 3 * rng.randn(50)
Xtr2 = np.concatenate((Xtr2_1, Xtr2_2), axis=0)
Xtr = np.column_stack((Xtr1, Xtr2))

ax = plt.axes()
ax.scatter(Xtr1_1, Xtr2_1, c = 'green');
ax.scatter(Xtr1_2, Xtr2_2, c = 'green');
plt.title("Training data")
```

```python
# Apply K-means clustering with K = n_clusters
kmeans = KMeans(n_clusters=4, random_state=0).fit(Xtr)
labels = kmeans.labels_
```

```python
print("Cluster Centers:")
kmeans.cluster_centers_
```

```python
plt.scatter(Xtr[:, 0], Xtr[:, 1], c=labels)
plt.title("K-means clustering results")
plt.xlabel("Xtr1")
plt.ylabel("Xtr2")
plt.show()
```
