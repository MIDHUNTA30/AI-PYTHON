$\large{\textbf{1.1 Artificial Intelligence (AI)}}$

*   $\color{red}{\textbf{AI}}$ deals with programming a computer so that it can learn from available input data and perform tasks involving intelligence:  understanding, thinking, and decision making.
*   Subfields of AI are
 1. $\color{red}{\textbf{Machine learning}}$ ($\textbf{ML}$): associated with learning models from data.
 2. $\color{red}{\textbf{Neural Network}}$ and $\color{red}{\textbf{Deep Learning}}$ ($\textbf{NN}$ and $\textbf{DL}$): deals with learning neural network based models from data.
 3. $\color{red}{\textbf{Natural Language Processing}}$ ($\textbf{NLP}$): deals with problems such as text recognition, speech recognition, language translation, information extraction and retrievel, etc.
 4. $\color{red}{\textbf{Robotics}}$: deals with design machines (robots) that can help or assist humans.
 5. $\color{red}{\textbf{Expert systems and control}}$: deals with planning and decision making (control) of processes, dynamical systems.
 6. $\color{red}{\textbf{Knowledge representation and reasoning}}$ ($\textbf{KRR}$): deals with  representing information about the world that can be used for solving complex problems.
 7. $\color{red}{\textbf{Social intelligence}}$: deals with developing algorithms that can recognize, interpret, and process human feelings and emotions.

$\large{\textbf{1.3 Types of AI algorithms}}$
1. $\color{red}{\textbf{Supervised learning}}$: used for obtaining model of a set of data containing the inputs and outputs ($\textbf{labels}$). Can be further divided as
      * Regression
      *Classification
2. $\color{red}{\textbf{Unsupervised learning}}$: used for obtaining model that extracts useful information
from the input ($\textbf{unlabelled data}$). Can be further divided as
      * Dimensionality reduction
      * Clustering
3. $\color{red}{\textbf{Reinforcement learning}}$: Is used to learn a sequence of decision or policy using data. Can be further divided as
      * Model-based algorithms
      * Model free algorithms
