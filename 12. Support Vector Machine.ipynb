$\large{\textbf{12.1 Support Vector Machine (SVM)}}$

* $\color{red}{\textbf{SVM}}$ is a linear classifier.
* SVM classify data by finding the best hyperplane.

* $\color{red}{\textbf{Hyperplane}}$: is the set of points $x \in \mathbb{R}^n$ satisfying

\begin{equation}
h(x)=Ax+b=0
\end{equation}

* The weight vector $A$ is orthogonal to the hyperplane.

*   Training data contains:

   *  $\color{red}{\textbf{Inputs}}$: consists of $N$ observations $\{ x_{1}, x_{2},...,x_{N} \}$
   where $x_{k} \in \mathbb{R}^n,$ $k=1,..,N$ are the $N$ input samples.
   *  $\color{red}{\textbf{Outputs}}$: consists of $N$ labels $\{ y_{1}, y_{2},...,y_{N} \}$
   where $y_{k} \in \{-1,+1\}$ are the class labels.

<center>

SVM1.png

</center>

* $\color{red}{\textbf{Seperating hyperplane}}$: splits the $n$ dimensional spaces into 2 halfspaces.

* Linearly seperable data: for which there exists a hyperplane such that each halfspace contains points from a single class.

* Let the labels be $y=-1$ for class 1 and $y=+1$ for class 2.

*  $\color{red}{\textbf{Marginal hyperplanes}}$: are two parallel hyperplanes that separate the classes, so that the distance between them is as large as possible.

* For a normalized dataset these hyperplanes can be defined as:
     * $Ax+b=-1$ : marginal for class 1
     * $Ax+b=+1$ : marginal for class 2
*  $\color{red}{\textbf{Support vectors}}$: are data points on the marginal hyperplanes.

* Class 1 ponts satisfy  $Ax+b \leq -1$ and class 2 satisfy $Ax+b\geq +1$. This can be combined as
\begin{equation}
  y_{k}(Ax_{k}+b) \geq 1,  \hspace{1cm}k=1,2,\dots,N.
\end{equation}

* The distance between these hyperplane is $\frac{2}{\parallel A \parallel}$. To maximize the distance, we have to minimize ${\parallel A \parallel}$.        


* Optimization problem for SVM:
\begin{equation}
\begin{aligned}
    \underset{A,b}{\min} \hspace{.5cm}   & {\parallel A \parallel}^{2} \\
    \text{subj. to}  ~&~ y_{k}(Ax_{k}+b) \geq 1,  \hspace{1cm}k=1,2,\dots,N.
\end{aligned}    
\end{equation}

$\large{\textbf{12.2 Illustrating SVM}}$

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
```

```python
rng = np.random.RandomState(1)
a=3
b=2

Xtr1_1 = np.linspace(-6, 2, 50)
Xtr1_2 = np.linspace(-2, 6, 50)
Xtr1=np.concatenate((Xtr1_1,Xtr1_2),axis=0)

Xtr2_1= a*Xtr1_1+5*b + 3*rng.randn(50)
Xtr2_2= a*Xtr1_2-5*b - 3*rng.randn(50)
Xtr2=np.concatenate((Xtr2_1,Xtr2_2),axis=0)
Xtr=(Xtr1,Xtr2)
Ytr_1=-1*np.ones(50)
Ytr_2=np.ones(50)
Ytr=np.concatenate((Ytr_1,Ytr_2),axis=0)
ax = plt.axes()
ax.scatter(Xtr1_1, Xtr2_1, c = 'green');
ax.scatter(Xtr1_2, Xtr2_2, c = 'red');
plt.title("Training data")
```

```python
model = svm.SVC(kernel='linear', degree=2)
model.fit(np.transpose(Xtr), Ytr)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

# Create grid to evaluate model
x1 = np.linspace(min(Xtr1) - 1, max(Xtr1) + 1, len(np.transpose(Xtr)))
x2 = np.linspace(min(Xtr2)-1, max(Xtr2) + 1, len(Ytr))
X1, X2 = np.meshgrid(x1, x2)
X = np.vstack([X1.ravel(), X2.ravel()]).T
train_size = len(Xtr1)
colors = np.where(Ytr == 1, 'red', 'green')  # Assigning different colors to classes
ax.scatter(Xtr1, Xtr2, c=colors)
Z = model.decision_function(X).reshape(X1.shape)
ax.contour(X1, X2, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])  # Plotting decision boundary and margins
ax.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], s=100, linewidth=1, facecolors='none', edgecolors='k')  # Highlighting support vectors
plt.show()
```
