$\large{\textbf{8.1 Neural Network}}$


*  NN is a network of neurons that are arranged in layers.
*  Each layer transforms its input data and gives an output.
*  Output of each layer will be the input of the next layer.
*  From a mathematical point of view, NN is a $\textbf{composite function}$ of
     the following form:

     \begin{equation}
       \hat{y}=f_{L+1}(f_{L}(...(f_{1}(f_{0}(x))))
     \end{equation}

     where $f_{0},f_{1},f_{2},...,f_{L+1}$ are $\textbf{layers}$ of the NN which are composite functions of the form:

\begin{equation}
    f_{i}(x)=\sigma(Ax+b)  
\end{equation}

*   The functions $f_{i}$ consist of two components:

  1. $\color{red}{\textbf{Affine function}}$: is the linear component which is of the form $Ax+b.$

  2.   $\color{red}{\textbf{Activation function}}$: is the nonlinear component. One of the popular activation function is the ReLU
      (Rectified Linear Unit) function which is defined as below:
        \begin{equation}
        \sigma(x)=\max(0,x)= \begin{cases}
        0, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        x, \hspace{0.3cm} &if \hspace{0.1cm} x >0
        \end{cases}
        \end{equation}

$\large{\textbf{8.2 Neurons and Layers}}$

${\textbf{Neuron}}$


*   $\color{red}{\textbf{Neurons}}$ are the building blocks of NN.
*   Neurons can be mathematically represented as
    \begin{equation}
    y=\sigma(Ax+b)
\end{equation}
* The input to the neuron or $x$ can be a scalar or a vector.
* The output to the neuron or $y$ will be a scalar.

${\textbf{Layer}}$


*   $\color{red}{\textbf{Layer}}$ is a collection of neurons that takes the same inputs.

*   Layers can be grouped into 3 categores:
     1.  $\color{red}{\textbf{Input layer}}$: in which each neuron has only one input and it simply passes the input. i.e. $A=1,$ $b=0,$ and $\sigma(Ax+b)=x.$
     2. $\color{red}{\textbf{Hidden layer}}$: in which each neuron can have more than one inputs and gives a scalar output. The activation function is normally noninear: ReLU, tanh, etc.
     2.  $\color{red}{\textbf{Output layer}}$: in which each neuron can have more than one inputs and  gives a scalar output. The activation function is normally linear.
  

$\large{\textbf{8.3. Activation functions}}$

*  Activation functions are the nonlinear component in an NN.

*   Activation functions that are commonly used in NNs are listed below:
 1.  $\color{red}{\textbf{Linear}}$ function:
      \begin{equation}
        \sigma(x)=ax
      \end{equation}
 2. $\color{red}{\textbf{ReLU}}$ (Rectified Linear Unit) function:
    \begin{equation}
        \sigma(x)=\max(0,x)= \begin{cases}
        0, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        x, \hspace{0.3cm} &if \hspace{0.1cm} x >0
        \end{cases}
    \end{equation}
 3. $\color{red}{\textbf{Parametric ReLU}}$ function:
    \begin{equation}
        \sigma(x)= \begin{cases}
        ax, \hspace{0.3cm} a>0, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        x, \hspace{0.3cm} &if \hspace{0.1cm} x >0
        \end{cases}
    \end{equation}   
 4. $\color{red}{\textbf{Modulus}}$ function:
    \begin{equation}
        \sigma(x)=|x|= \begin{cases}
        -x, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        x, \hspace{0.3cm} &if \hspace{0.1cm} x >0
        \end{cases}
    \end{equation}   

 5. $\color{red}{\textbf{Step}}$ function:
     \begin{equation}
        \sigma(x)=\begin{cases}
        0, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        1, \hspace{0.3cm} &if \hspace{0.1cm} x >0
        \end{cases}
     \end{equation}
 6. $\color{red}{\textbf{Signum}}$ function:
     \begin{equation}
        \sigma(x)=\begin{cases}
        -1, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        1, \hspace{0.3cm} &if \hspace{0.1cm} x >0
        \end{cases}
     \end{equation}    

 7. $\color{red}{\textbf{Saturation}}$ (Satlin) function:
     \begin{equation}
        \sigma(x)=\begin{cases}
        -1, \hspace{0.3cm} & if \hspace{0.1cm} x\leq -1\\
        x, \hspace{0.3cm} & if \hspace{0.1cm} -1< x < 1\\
        1, \hspace{0.3cm} &if \hspace{0.1cm} x \geq 1
        \end{cases}
     \end{equation}

 8. $\color{red}{\textbf{Sigmoid}}$ function:
    \begin{equation}
      \sigma(x)=\frac{1}{1+e^{-x}}
    \end{equation}
 9.   $\color{red}{\textbf{Swish}}$  function:
    \begin{equation}
        \sigma(x)= x\hspace{.1cm} sigmoid(\beta x)=\frac{x}{1+e^{-\beta x}}
    \end{equation}

 10.   $\color{red}{\textbf{Tanh}}$ (hyperbolic tan) function:
    \begin{equation}
        \sigma(x)=\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}
    \end{equation}



*  Consider an NN with one hidden layer and ReLU activation function. The number of neurons in the hidden layer $m_{1}$ is selected as 100 and the weights and biases for each neuron are selected randomly.

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
m1=100
X = np.linspace(-10, 10, 100)
Y1=np.zeros(m1)
Y=np.zeros(len(X))
A1=np.random.randn(m1)        # weights for 100 neurons are selected randomly
b1=np.random.randn(m1)     # biases for 100 neurons are selected randomly
for i in range(len(X)):
 for j in range(m1):
  Y1[j] = np.max([(A1[j]*X[i]+b1[j]),0])
 Y[i]  = np.sum(Y1)
plt.figure(figsize =(6, 4))
plt.plot(X, Y, c = 'red');
```

 *   The output values will be always positive in this case.

 *   This is because of the reason that here we defined the output as the sum of elements in the array $Y1,$ which is equivalent to the output weight $A2=[1 \hspace{.3cm} 1 \hspace{.3cm}\dots \hspace{.3cm} 1]$ which contains 100 ones and bias $b=0.$
 Further, each element of Y1 will be nonnegative since they are outputs of a ReLU function.
 *  To get negative outputs, we can include negative values in the output weight $A2.$ This is illustrated next, in which the output weight $A2$ is chosen as a random array with 100 elements.

```python
m1=100
X = np.linspace(-10, 10, 100)
Y1=np.zeros(m1)
Y=np.zeros(len(X))
A1=np.random.randn(m1)
b1=np.random.randn(m1)
A2=np.random.randn(m1)
b2=np.random.randn(1)
for i in range(len(X)):
 for j in range(m1):
  Y1[j] = np.max([(A1[j]*X[i]+b1[j]),0])
 Y[i]  = np.sum(np.multiply(A2,Y1)+b2)
plt.figure(figsize =(6, 4))
plt.plot(X, Y, c = 'red');
```

*  In the above plot we can observe that the output values are negative. Here if we rerun the code we can get both positive and negative output values depending on the elements in the weight matrix A2.

$\textbf{Modulus function}$

* The graph of the modulus function can be plotted using the code below:

```python
X = np.linspace(-10, 10, 100)
Y=np.zeros(len(X))
for i in range(len(X)):
 Y[i] = np.abs(X[i])
plt.figure(figsize =(6, 4))
plt.plot(X, Y, c = 'red');
```

* Now consider an NN with one hidden layer which has 100 neurons with modulus activation function.

```python
m1=100
X = np.linspace(-10, 10, 100)
Y1=np.zeros(m1)
Y=np.zeros(len(X))
A1=np.random.randn(m1)
b1=np.random.randn(m1)
A2=np.random.randn(m1)
b2=np.random.randn(1)
for i in range(len(X)):
 for j in range(m1):
  Y1[j] = np.abs(A1[j]*X[i]+b1[j])
 Y[i]  = np.sum(np.multiply(A2,Y1)+b2)
plt.figure(figsize =(6, 4))
plt.plot(X, Y, c = 'red');
```

* Here we can see that the output is piecewise linear like the ReLU function. This is because modulus function is a piecewise linear function

$\textbf{Step function}$

* We can plot the graph of the step function using the code below:

```python
X = np.linspace(-10, 10, 100)
Y=np.zeros(len(X))
for i in range(len(X)):
 Y[i] = np.heaviside(X[i],0)
plt.figure(figsize =(6, 4))
plt.plot(X, Y, c = 'red');
```

* Now consider an NN with one hidden layer which has 100 neurons with step activation function.

```python
m1=100
X = np.linspace(-10, 10, 100)
Y1=np.zeros(m1)
Y=np.zeros(len(X))
A1=np.random.randn(m1)
b1=np.random.randn(m1)
A2=np.random.randn(m1)
b2=np.random.randn(1)
for i in range(len(X)):
 for j in range(m1):
  Y1[j] = np.heaviside(A1[j]*X[i]+b1[j],0)
 Y[i]  = np.sum(np.multiply(A2,Y1)+b2)
plt.figure(figsize =(6, 4))
plt.plot(X, Y, c = 'red');
```

* Here we can see that the output is piecewise linear. This is because of the reason that step function is a piecewise linear function. However here we can observe that the folds in the line are harder compared to the one with ReLU function.

$\textbf{Sigmoid function}$

* We can plot the graph of the sigmoid function using the code below:

```python
X = np.linspace(-10, 10, 100)
Y=np.zeros(len(X))
for i in range(len(X)):
 Y[i] = np.divide(1, 1+ np.exp(-(X[i])))  # implemented sigmoid(X[i])
plt.figure(figsize =(6, 4))
plt.plot(X, Y, c = 'red');
```

* Now consider an NN with one hidden layer which has 100 neurons with sigmoid activation function.

```python
m1=100
X = np.linspace(-10, 10, 100)
Y1=np.zeros(m1)
Y=np.zeros(len(X))
A1=np.random.randn(m1)
b1=np.random.randn(m1)
A2=np.random.randn(m1)
b2=np.random.randn(1)
for i in range(len(X)):
 for j in range(m1):
  Y1[j] = np.divide(1, 1+ np.exp(-(A1[j]*X[i]+b1[j])))  # implemented sigmoid(A1[j]*X[i]+b1[j])
 Y[i]  = np.sum(np.multiply(A2,Y1)+b2)
plt.figure(figsize =(6, 4))
plt.plot(X, Y,  c = 'red');
```

* The output is a smooth curve. This is because of the continuously differentiable nature of the sigmoid function.

$\textbf{Swish function}$

* We can plot the graph of the swish function using the code below:

```python
X = np.linspace(-10, 10, 100)
Y=np.zeros(len(X))
beta=0.5
for i in range(len(X)):
 Y[i] = np.divide(X[i], 1+ np.exp(-(beta*X[i])))  # implemented swish(X[i])
plt.figure(figsize =(6, 4))
plt.plot(X, Y, c = 'red');
```

* Now consider an NN with one hidden layer which has 100 neurons with swish activation function.

```python
m1=100
X = np.linspace(-10, 10, 100)
Y1=np.zeros(m1)
Y=np.zeros(len(X))
A1=np.random.randn(m1)
b1=np.random.randn(m1)
A2=np.random.randn(m1)
b2=np.random.randn(1)
for i in range(len(X)):
 for j in range(m1):
  Y1[j] = np.divide((A1[j]*X[i]+b1[j]), 1+ np.exp(-(beta*(A1[j]*X[i]+b1[j]))))  # implemented swish(A1[j]*X[i]+b1[j])
 Y[i]  = np.sum(np.multiply(A2,Y1)+b2)
plt.figure(figsize =(6, 4))
plt.plot(X, Y,  c = 'red');
```

```python
m1=100
X = np.linspace(-10, 10, 100)
Y1=np.zeros(m1)
Y=np.zeros(len(X))
A1=np.random.randn(m1)
b1=np.random.randn(m1)
A2=np.random.randn(m1)
b2=np.random.randn(1)
for i in range(len(X)):
 for j in range(m1):
  Y1[j] = np.tanh(A1[j]*X[i]+b1[j])
 Y[i]  = np.sum(np.multiply(A2,Y1)+b2)
plt.figure(figsize =(6, 4))
plt.plot(X, Y, c = 'red');
```

$\textbf{Comparing activation functions}$



*   First let's compare the graph of different activation functions.



```python
m1=100
X = np.linspace(-10, 10, 100)
Ya=np.zeros(len(X))
Yb=np.zeros(len(X))
Yc=np.zeros(len(X))
Yd=np.zeros(len(X))
Ye=np.zeros(len(X))
Yf=np.zeros(len(X))
Yg=np.zeros(len(X))
Yh=np.zeros(len(X))
for i in range(len(X)):
  Ya[i] = np.max([X[i],0])
  Yb[i] = np.abs(X[i])
  Yc[i] = np.heaviside(X[i],0)
  Yd[i] = np.sign(X[i])
  Ye[i] = np.clip(X[i],-1,1)
  Yf[i] = np.divide(1, 1+ np.exp(-X[i]))
  Yg[i] = np.divide(X[i], 1+ np.exp(-beta*X[i]))
  Yh[i] = np.tanh(X[i])
fig, axs = plt.subplots(2, 4, figsize =(12, 6))
axs[0, 0].plot(X, Ya, c = 'r');
axs[0, 0].set_title("ReLU")
axs[0, 1].plot(X, Yb, c = 'r');
axs[0, 1].set_title("Modulus")
axs[0, 2].plot(X, Yc, c = 'r');
axs[0, 2].set_title("Step")
axs[0, 3].plot(X, Yd, c = 'r');
axs[0, 3].set_title("Signum")
axs[1, 0].plot(X, Ye, c = 'r');
axs[1, 0].set_title("Saturation")
axs[1, 1].plot(X, Yf, c = 'r');
axs[1, 1].set_title("Sigmoid")
axs[1, 2].plot(X, Yg, c = 'r');
axs[1, 2].set_title("Swish")
axs[1, 3].plot(X, Yh, c = 'r');
axs[1, 3].set_title("Tanh")
plt.show()
```

*   The following code compares the output of an NN with different activation functions.

```python
m1=100
X = np.linspace(-10, 10, 100)
Y1a=np.zeros(m1)
Y1b=np.zeros(m1)
Y1c=np.zeros(m1)
Y1d=np.zeros(m1)
Y1e=np.zeros(m1)
Y1f=np.zeros(m1)
Y1g=np.zeros(m1)
Y1h=np.zeros(m1)
Ya=np.zeros(len(X))
Yb=np.zeros(len(X))
Yc=np.zeros(len(X))
Yd=np.zeros(len(X))
Ye=np.zeros(len(X))
Yf=np.zeros(len(X))
Yg=np.zeros(len(X))
Yh=np.zeros(len(X))
A1=np.random.randn(m1)
b1=np.random.randn(m1)
A2=np.random.randn(m1)
b2=np.random.randn(1)
for i in range(len(X)):
 for j in range(m1):
  Y1a[j] = np.max([(A1[j]*X[i]+b1[j]),0])
  Y1b[j] = np.abs(A1[j]*X[i]+b1[j])
  Y1c[j] = np.heaviside(A1[j]*X[i]+b1[j],0)
  Y1d[j] = np.sign(A1[j]*X[i]+b1[j])
  Y1e[j] = np.clip(A1[j]*X[i]+b1[j],-1,1)
  Y1f[j] = np.divide(1, 1+ np.exp(-(A1[j]*X[i]+b1[j])))
  Y1g[j] = np.divide((A1[j]*X[i]+b1[j]), 1+ np.exp(-(beta*(A1[j]*X[i]+b1[j]))))
  Y1h[j] = np.tanh(A1[j]*X[i]+b1[j])
 Ya[i]  = np.sum(np.multiply(A2,Y1a)+b2)
 Yb[i]  = np.sum(np.multiply(A2,Y1b)+b2)
 Yc[i]  = np.sum(np.multiply(A2,Y1c)+b2)
 Yd[i]  = np.sum(np.multiply(A2,Y1d)+b2)
 Ye[i]  = np.sum(np.multiply(A2,Y1e)+b2)
 Yf[i]  = np.sum(np.multiply(A2,Y1f)+b2)
 Yg[i]  = np.sum(np.multiply(A2,Y1g)+b2)
 Yh[i]  = np.sum(np.multiply(A2,Y1h)+b2)

fig, axs = plt.subplots(2, 4, figsize =(12, 6))
axs[0, 0].plot(X, Ya, c = 'r');
axs[0, 0].set_title("ReLU")
axs[0, 1].plot(X, Yb, c = 'r');
axs[0, 1].set_title("Modulus")
axs[0, 2].plot(X, Yc, c = 'r');
axs[0, 2].set_title("Step")
axs[0, 3].plot(X, Yd, c = 'r');
axs[0, 3].set_title("Signum")
axs[1, 0].plot(X, Ye, c = 'r');
axs[1, 0].set_title("Saturation")
axs[1, 1].plot(X, Yf, c = 'r');
axs[1, 1].set_title("Sigmoid")
axs[1, 2].plot(X, Yg, c = 'r');
axs[1, 2].set_title("Swish")
axs[1, 3].plot(X, Yh, c = 'r');
axs[1, 3].set_title("Tanh")
plt.show()
```
